Exemplo clássico de uso do Keras Applications para fazer classificação de imagens usando a ResNet50 já treinada no ImageNet.

```python
# 1. Imports

from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

ResNet50 → Modelo de rede neural profunda com 50 camadas, pré-treinada no dataset ImageNet (1,2 milhão de imagens, 1000 classes).

image → Módulo do Keras para carregar e preparar imagens.

preprocess_input → Função que normaliza os pixels da imagem no formato que a ResNet50 espera.

decode_predictions → Converte a saída numérica da rede (vetor de probabilidades) para rótulos de classes legíveis.

numpy → Para manipulação de arrays.

```python
# 2. Carregando o modelo pré-treinado

model = ResNet50(weights='imagenet')

```

Baixa automaticamente os pesos treinados no ImageNet.

O modelo está pronto para inferência (não precisa treinar).

```python
# 3. Carregando e preparando a imagem

img_path = '/content/Image001.jpg'
img = image.load_img(img_path, target_size=(224, 224))
```

Abre a imagem Image001 (Imagem de exemplo

Redimensiona para 224×224 pixels, que é o tamanho de entrada padrão da ResNet50.

```python
# Converte a imagem para um array NumPy no formato (224, 224, 3) com valores de 0 a 255.

x = image.img_to_array(img)
```

```python
# Adiciona uma dimensão para representar o batch → formato final (1, 224, 224, 3).

# Isso é necessário pois o modelo espera processar lotes de imagens, mesmo que haja apenas uma.

x = np.expand_dims(x, axis=0)
```

```python
# Normaliza os valores de pixels (subtraindo médias, mudando escala, etc.) para o formato que a ResNet50 foi treinada.

x = preprocess_input(x)
```

```python
# 4. Fazendo a predição

preds = model.predict(x)
```

Passa a imagem pelo modelo.

O resultado é um vetor de tamanho (1, 1000) com a probabilidade para cada classe do ImageNet.

```python
# 5. Decodificando a predição

print('Predicted:', decode_predictions(preds, top=3)[0])
```

Converte as 3 classes mais prováveis para nomes e probabilidades.

Esse código pega uma imagem, pré-processa, passa pela ResNet50 pré-treinada e retorna as classes mais prováveis. É um pipeline clássico de inference em transfer learning.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Carrega a imagem
img = mpimg.imread('/content/Image001.jpg')

# Exibe a imagem
plt.imshow(img)
plt.axis('off')  # Remove os eixos para melhor visualização
plt.show()
```

```python
# Caminho da imagem
img_path = '/content/Image001.jpg'
```

```python
# Carregar e pré-processar a imagem
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)  # adiciona dimensão batch
x = preprocess_input(x)
```

```python
# Fazer a predição
preds = model.predict(x)
```

```python
# Decodificar as 3 classes mais prováveis
print('Predicted:', decode_predictions(preds, top=3)[0])
```
